Задание 1 — множество уникальных геометок

```python
ids = {
    'user1': [213, 213, 213, 15, 213],
    'user2': [54, 54, 119, 119, 119],
    'user3': [213, 98, 98, 35]
}

unique_geo = set()

for marks in ids.values():
    unique_geo.update(marks)

print(unique_geo)
```

Задание 2 — распределение количества слов в запросах

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

total = len(queries)

stats = {}

for q in queries:
    word_count = len(q.split())
    stats[word_count] = stats.get(word_count, 0) + 1

for count, qty in sorted(stats.items()):
    percent = qty / total * 100
    print(f"Поисковых запросов, содержащих {count} слов(а): {percent:.2f}%")
```

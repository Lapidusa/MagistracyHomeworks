```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'},
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def get_owner_by_number(doc_number: str, documents_list: list[dict]) -> str | None:
    for doc in documents_list:
        if doc.get('number') == doc_number:
            return doc.get('name')
    return None


def get_shelf_by_number(doc_number: str, dirs: dict[str, list[str]]) -> str | None:
    for shelf, docs in dirs.items():
        if doc_number in docs:
            return shelf
    return None


def handle_command_p():
    doc_number = input("Введите номер документа:\n").strip()
    owner = get_owner_by_number(doc_number, documents)
    if owner is not None:
        print(f"Владелец документа: {owner}")
    else:
        print("Документ с таким номером не найден.")


def handle_command_s():
    doc_number = input("Введите номер документа:\n").strip()
    shelf = get_shelf_by_number(doc_number, directories)
    if shelf is not None:
        print(f"Документ хранится на полке: {shelf}")
    else:
        print("Документ с таким номером не найден.")


def main_loop():
    while True:
        command = input("Введите команду (p - владелец, s - полка, q - выход):\n").strip().lower()
        if command == 'q':
            print("Выход из программы.")
            break
        elif command == 'p':
            handle_command_p()
        elif command == 's':
            handle_command_s()
        else:
            print("Неизвестная команда. Доступные команды: p, s, q.")

main_loop()
```

Задание 1 — средняя буква или две средние

```python
word = 'testing'

length = len(word)

if length % 2 == 1:
    middle_index = length // 2
    result = word[middle_index]
else:
    middle_left = length // 2 - 1
    middle_right = length // 2
    result = word[middle_left : middle_right + 1]

print(result)
```

Задание 2 — пары из dating-сервиса

```python
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

boys_sorted = sorted(boys)
girls_sorted = sorted(girls)

if len(boys_sorted) != len(girls_sorted):
    print("Внимание, кто-то может остаться без пары!")
else:
    print("Идеальные пары:")
    for boy, girl in zip(boys_sorted, girls_sorted):
        print(f"{boy} и {girl}")
```
